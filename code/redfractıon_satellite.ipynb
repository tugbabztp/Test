```python
mask = (data["StarFormationRate"]!=0)
array = data.as_matrix()
array = array[mask]
array[:,10] = np.log10(array[:,10]/array[:,9]) #biz burdan log SSFR
array[:,9] = np.log10(array[:,9]) #buda logMstar
median_logssfr = np.median(array[:,10])

def central_filterer(logmmin, logmmax):
    """
    this function 
    keeps the central galaxies with non-zero SFR with logmstar 
    between logmmin, logmmax
    
    return:
    all galaxies with the above properties
    red galaxies with the baove properties
    blue galaxies with the above properties
    """
    
    central_mask = (data["SubGroupNumber"]==0)&(data["StarFormationRate"]!=0)
    array = data.as_matrix()
    array = array[central_mask]
    array[:,10] = np.log10(array[:,10]/array[:,9]) #biz burdan log SSFR
    array[:,9] = np.log10(array[:,9]) #buda logMstar
    
    mstar_mask = (array[:,9]>logmmin)&(array[:,9]<logmmax)
    
    array= array[mstar_mask,:]
    
    red_mask = array[:,10] < median_logssfr
    blue_mask = array[:,10] > median_logssfr
        
    red = array[red_mask,:]
    blue = array[blue_mask,:]
    tot= array 
    
    return tot,red, blue
    
    
def sat_filterer(logmmin, logmmax):
    """
    this function 
    keeps the central galaxies with non-zero SFR with logmstar 
    between logmmin, logmmax
    
    return:
    all galaxies with the above properties
    red galaxies with the baove properties
    blue galaxies with the above properties
    """
    
    sat_mask = (data["SubGroupNumber"]!=0)&(data["StarFormationRate"]!=0)
    array = data.as_matrix()
    array = array[sat_mask]
    array[:,10] = np.log10(array[:,10]/array[:,9]) #biz burdan log SSFR
    array[:,9] = np.log10(array[:,9]) #buda logMstar
    
    mstar_mask = (array[:,9]>logmmin)&(array[:,9]<logmmax)
    
    array= array[mstar_mask,:]
        
    red_mask = array[:,10] < median_logssfr
    blue_mask = array[:,10] > median_logssfr
    
        
    red = array[red_mask,:]
    blue = array[blue_mask,:]
    tot= array 
    
    return tot,red, blue
        
```

```python
def modify(x):
        
    mask = np.abs(x) > 50
    
    x[mask] = 100 - np.abs(x[mask])
    
    return np.abs(x)


def func(amin,bmax,data1,data2):
    
    """this function returns average number of 
    data1 points and data2 points between amin and bmax"""
    
    zebra = []

    for point in data1:
        
        matrix = point[None,:]-data2
        matrix = modify(matrix)
        
        distance_to_point = np.sum((matrix)**2  ,  axis = 1 )**0.5 #distance from all the points in fake data to the point
        A = (amin < distance_to_point) & ( distance_to_point < bmax)
        x = len(distance_to_point[A])
        zebra.append(x)
    zebra= np.array(zebra)
    return np.average(zebra), np.std(zebra)
```

```python
def func2(logmin,logmmax, galaxy_type1, galaxy_type2):
    """
    it returns func at (2,4),(4,6) etc. sonun da (2,4) için net sonucu verir.  """
    
    tot, red, blue = central_filterer(logmin, logmmax)

    
    
    if galaxy_type1 == 'red':
        FAKE_DATA1 = red[:,4:7] #bu demek ki biz tüm kolonı tutuyoruz ve sadece 4-7 arasındakı kolonları seçiyoruz.
    if galaxy_type1 == 'blue':
        FAKE_DATA1 = blue[:,4:7]
    if galaxy_type1 == 'tot':
        FAKE_DATA1 = tot[:,4:7]
   
    tot, red, blue = sat_filterer(logmin, logmmax)


    if galaxy_type2 == 'red':
        FAKE_DATA2 = red[:,4:7]
    if galaxy_type2 == 'blue':   #sns.distplot(array[:,10], norm_hist=False,kde=False)
        FAKE_DATA2 = blue[:,4:7]
    if galaxy_type2 == 'tot':
        FAKE_DATA2 = tot[:,4:7]
    
    list_average = []
    list_std = []
    
    lister = [0.1 + i*0.1 for i in range(20)]
    for i in lister:
        average , std = func(i,i+1,FAKE_DATA1,FAKE_DATA2)
        list_average.append(average)
        list_std.append(std)
    return np.array(list_average), np.array(list_std)
```

```python
logmmin = 9.5
logmmax = 9.7


tot_red , tot_red_err = func2(logmmin,logmmax,'red','tot')
red_red , red_red_err = func2(logmmin,logmmax,'red','red')
tot_blue, tot_blue_err = func2(logmmin,logmmax,'blue','tot')
red_blue, red_blue_err = func2(logmmin,logmmax,'blue','red')

fred_red = np.array(red_red)*1.  / np.array(tot_red) #equation 1 of paper
fred_blue = np.array(red_blue)  / np.array(tot_blue) #equatıon 2 of paper

var_fred_red = (red_red_err**2/red_red**2 + tot_red_err**2/tot_red**2)*(red_red/tot_red)**2.  
fred_red_err = var_fred_red**.5

plt.figure(figsize=(10,10))

x =  [0.15 + i*0.1 for i in range(20)]
y1 = fred_red
y2 = fred_blue

#plt.plot(x,red_red)
#plt.plot(x,tot_red)
plt.plot(x,fred_red, marker = "o", color ="r" ,lw =3,  label = '$P(sat = red | central = red )$')
plt.plot(x,y2, marker = "o", color ="b",lw =3 , label= '$P(sat = red | central = blue )$')
plt.title(str(logmmin)+'$<logM_{star}<$'+str(logmmax), fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.ylabel("the red fraction",fontsize=25)
plt.xlabel("distance (Mpc)",fontsize=25)
plt.legend(fontsize = 25)
plt.ylim([0,1])
plt.savefig('fred'+str(logmmax)+'.pdf')
plt.show()
plt.close()

plt.figure(figsize=(10,10))
plt.plot(x,y1-y2, marker = "o", color ="k", lw =3 )
plt.axhline(0, linestyle='dashed', color = "r") #ıkıncı grafıktekı zerodaki uzun çizgiyi çizmemize yardımcı oluyor.
plt.title(str(logmmin)+r'$<logM_{\rm star}<$'+str(logmmax), fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.ylabel(r'$\Delta f_{\rm red}$',fontsize=25)
plt.xlabel("distance (Mpc)",fontsize=25)
plt.ylim([-0.5,0.5])
plt.savefig('dfred'+str(logmmax)+'.pdf')
```
